```python
#importing libraries needed
import numpy as np
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#gold labels taken from the test data label column. Zero and one shot labels are obtained as decsribed in Step 5 of the project report
gold_labels = ['B1', 'A1', 'C1', 'A1', 'A1', 'A2', 'A2', 'A1', 'B1', 'A1', 'B1', 'B1', 'C2', 'A2', 'B1', 'C2', 'B2', 'B2', 'B2', 'C1', 'B2', 'C1', 'C1', 'C1', 'C2','A2', 'C2', 'C2', 'A2', 'B2']

zero_shot_labels = ['B1', 'A2', 'C2', 'A1', 'B1', 'A2', 'B1', 'A1', 'B2', 'A1', 'B2', 'B1', 'C1', 'A2', 'B1', 'C1', 'C1', 'B2', 'B2', 'C1', 'B1', 'C1', 'B2', 'B2', 'B1', 'A2', 'C1', 'C1', 'A1', 'B2']

one_shot_labels = ['A2', 'A1', 'C1', 'A2', 'B1', 'A2', 'B1', 'A1', 'B2', 'A1', 'B2', 'B1', 'C1', 'A2', 'B1', 'C1','B2', 'C2', 'C1', 'C2', 'B1', 'C1', 'C2', 'C1', 'B2', 'A2', 'B2', 'C1', 'A1', 'C1']
```

```python
#checking for skipped labels
print(len(gold_labels))
print(len(zero_shot_labels))
print(len(one_shot_labels))
```

```python
#evaluation of zero shot labels
class_report = classification_report(gold_labels, zero_shot_labels)
print(class_report)
```

```python

conf_matrix = confusion_matrix(gold_labels, zero_shot_labels)
print('Confusion Matrix:\n', conf_matrix)
print('Classification Report:\n', class_report)

# Plot the confusion matrix
import seaborn as sns
xticklabels = ['A1', 'A2', 'B1', 'B2', 'C1', 'C2']
yticklabels = ['A1', 'A2', 'B1', 'B2', 'C1', 'C2']
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=xticklabels, yticklabels=yticklabels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Zero-Shot Predicted Labels')
plt.show()
```

```python
#Evaluating the one-shot labels
class_report_2 = classification_report(gold_labels, one_shot_labels)
print(class_report_2)
```

```python
#Confusion matrix
conf_matrix_2 = confusion_matrix(gold_labels, one_shot_labels)

print('Confusion Matrix:\n', conf_matrix_2)
print('Classification Report:\n', class_report_2)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_2, annot=True, fmt='d', cmap='Blues', xticklabels=xticklabels, yticklabels=yticklabels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for One-Shot Predicted Labels')
plt.show()
```
